```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv("Housing.csv")
```

```python
print(df)
```

```python
profile = ProfileReport(df)
profile.to_file(output_file = "housing_report.html")
```

```python
plt.figure(figsize = (20, 10))
sns.regplot(data = df, y = "area", x = "price", color = "c").set(title = "area vs price")
```

```python
plt.figure(figsize = (20, 10))
sns.barplot(x = "bedrooms", y = "area", data = df)
plt.title("number of bedrooms and area")
```
